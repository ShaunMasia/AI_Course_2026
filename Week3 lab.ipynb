# <u> Part 1: Numpy (The Math Engine) </u>

```python
import numpy as np

#Creating a Matrix ( 2 rows , 3 columns )
A = np.array([
    [1,2,3],
    [4,5,6]])

print ("Matrix Shape:", A.shape)

#Reshaping: Flattening a 2x3 matrix into a (1x6) vector
A_Reshape = A.reshape(1,6)
print("Reshaped:", A_Reshape)

#Simple Broadcasting Operation
prices = np.array([100,200,300])
discounted_prices = prices * 0.9
print("Discounted Prices:", discounted_prices)
```

#### Machine Learning Essentials: Matrix Math & Logic

```python
#Inputs (3 features)
inputs = np.array([1.0,2.0,3.0])

#Weights (What the AI learns)
weights = np.array([0.2,0.8,-0.5])

#Bias (Offset) "Seasoning"
bias = 2.0

#Calculation: (Inputs * Weights) + bias = Neuron Output
output = np.dot(inputs,weights) + bias
print("Neuron Output:", output)

#Prediction: if Output > 2.5 , its 'malicious' else 'safe'
prediction = "Malicious" if output > 2.5 else "safe"
print("Model_Prediction:", prediction)

#Evaluation: Calculating Mean Absolute Error (MAE)
ground_truth = np.array([3.0])
loss = np.mean(np.abs(output - ground truth))

```

# <u>Part 2: Pandas (Data Manipulation)

```python
import pandas as pd

#Creating a dataframe
data= {"Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
"Age": [24, 27, 22, 32, 29],
"Score": [85, 90, 88, 92, 95],
"City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
}
df=pd.DataFrame(data)
print(df)
```

```python
# Get statistical summary (Mean, Max, Min)
print("\n--- Describe ---")
print(df.describe())

#Filter data: Who scored above 90?
print("\n--- High Scorers ---")
high_scores = df[df["Score"]>90]
print(high_scores)
```

### Handling Missing Data

```python
#Simulating missing data
df.loc[2,"Score"]= np.nan # Charlie lost his score
print(df)
```

```python
# Strategy 1: Drop the row (If you have lots of data)
df_clean = df.dropna()
print("\nDropped NaNs:\n", df_clean)
```

```python
# Strategy 2: Fill the hole (Imputation)
# Fill with the average score of everyone else
mean_score = df["Score"].mean()
df["Score"] = df["Score"].fillna(mean_score)
print("\nFilled NaNs:\n",df)
```

# <u>Part 3: Matplotlib (Visualization) </u>

```python
import matplotlib.pyplot as plt

x= np.linspace(0, 10,100) # 100 numbers from 0 to 10
y = np.sin(x)             #The Sine wave

plt.figure(figsize=(10,6))
plt.plot(x,y, label ='sine wave', color='blue')
plt.title("My First AI plot")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
